```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/internship projects/train.csv")
print(data)
```

```python
print("First few rows of the dataset:")
print(data.head(10))
```

```python

print("\nSummary Statistics:")
print(data.describe())
```

CHECKING A MISSING VALUES



```python
print(data.isnull().sum())
```

```python
# Visualize missing values
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title("Heatmap of Missing Values")
plt.show()
```

# Fill missing values for 'Age' with median and drop 'Cabin' due to many missing values

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data.drop(columns=['Cabin'], inplace=True)
```

```python
data.nunique()
```

```python
data.notnull()
```

FINALLY TO HANDLING A MISSING VALUES AND THERE IS NO NULL VALUES OBTAINED IN THE DATASETS, AFTER DATA CLENSING PROCESS



RELATIONSHIP BETWEEN THE VARIABLES


```python
sns.countplot(data=data, x='Survived', palette='Set2')
plt.title("Survival Count")
plt.show()

sns.histplot(data['Age'], kde=True, bins=30, color='blue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Bivariate Analysis
sns.countplot(data=data, x='Survived', hue='Sex', palette='coolwarm')
plt.title("Survival by Gender")
plt.show()

sns.countplot(data=data, x='Survived', hue='Pclass', palette='viridis')
plt.title("Survival by Passenger Class")
plt.show()

sns.boxplot(data=data, x='Survived', y='Age', palette='pastel')
plt.title("Age Distribution by Survival")
plt.show()



# 5. Fare Distribution by Class
sns.boxplot(data=data, x='Pclass', y='Fare', palette='muted')
plt.title("Fare Distribution by Passenger Class")
plt.show()

# 6. Survival by Embarkation Port
sns.countplot(data=data, x='Embarked', hue='Survived', palette='husl')
plt.title("Survival by Embarkation Port")
plt.show()

# 7. Pairplot
sns.pairplot(data[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='coolwarm')
plt.title("Pairplot of Key Features")
plt.show()

# 8. Check final dataset shape
print("\nFinal Shape of the Dataset:", data.shape)

# Conclusion
print("\nEDA Completed! Key insights should now be visible from the visualizations.")
```
