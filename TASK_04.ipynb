```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Load the dataset
data = pd.read_csv("/content/drive/MyDrive/internship projects/cleaned.csv")

# Display the first few rows of the dataset
print("Dataset Preview:")
data.head()

# Check for missing values and data types
print("\nData Info:")
data.info()

# Descriptive statistics
print("\nDescriptive Statistics:")
print(data.describe())

# 1. Analyzing the impact of road surface type, weather conditions, and light conditions

# Plotting accident counts by Road Surface Type
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Road_surface_type', order=data['Road_surface_type'].value_counts().index)
plt.title("Accidents by Road Surface Type")
plt.xlabel("Road Surface Type")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()

# Plotting accident counts by Weather Conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Weather_conditions', order=data['Weather_conditions'].value_counts().index)
plt.title("Accidents by Weather Conditions")
plt.xlabel("Weather Conditions")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()

# Plotting accident counts by Light Conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Light_conditions', order=data['Light_conditions'].value_counts().index)
plt.title("Accidents by Light Conditions")
plt.xlabel("Light Conditions")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()



# Investigate specific factors (e.g., Accident Severity vs Weather Conditions)
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Weather_conditions', y='Accident_severity')
plt.title("Accident Severity by Weather Conditions")
plt.xlabel("Weather Conditions")
plt.ylabel("Accident Severity")
plt.xticks(rotation=45)
plt.show()

# 3. Visualizing other categorical factors (e.g., Driving Experience and Accident Severity)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Driving_experience', hue='Accident_severity', order=data['Driving_experience'].value_counts().index)
plt.title("Accident Severity by Driving Experience")
plt.xlabel("Driving Experience")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

```python

# 4. Visualizing hotspots using a heatmap
plt.figure(figsize=(12, 8))
agg_data = data.groupby(['Weather_conditions', 'Light_conditions']).size().reset_index(name='Accident Count')
hotspot_data = agg_data.pivot(index='Weather_conditions', columns='Light_conditions', values='Accident Count')
sns.heatmap(hotspot_data, annot=True, cmap="YlOrRd", fmt=".0f")
plt.title("Accident Hotspots by Weather and Light Conditions")
plt.xlabel("Light Conditions")
plt.ylabel("Weather Conditions")
plt.show()

# 5. Visualizing contributions of factors to accident severity
factor_contributions = data.groupby('Accident_severity').size().reset_index(name='Count')
plt.figure(figsize=(8, 6))
sns.barplot(data=factor_contributions, x='Accident_severity', y='Count', palette="viridis")
plt.title("Contributions to Accident Severity")
plt.xlabel("Accident Severity")
plt.ylabel("Count")
plt.show()
```
